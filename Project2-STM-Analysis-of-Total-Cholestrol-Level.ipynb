# Analysis of Total Cholestrol Level

The program analyzes spread of total chlolestrol level readings from mean value. A greater deviation from mean values greater risk of heart attack which is seen for the group leading unhealthy and unregulated life. Where as the other group who do regular exercises and lead a controlled life have less fluctuations in his cholestrol level.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for Total Cholestrol Level
np.random.seed(0)  # For reproducibility

# Patient Group 1: Low Standard Deviation
mean_bp1 = 180  # Mean Total Cholestrol Level
std_dev1 = 5    # Low standard deviation
bp_readings1 = np.random.normal(mean_bp1, std_dev1, 100)

# Patient Group 2: High Standard Deviation
mean_bp2 = 200  # Mean Cholestrol Level
std_dev2 = 30   # High standard deviation
bp_readings2 = np.random.normal(mean_bp2, std_dev2, 100)



# Plotting the data
plt.figure(figsize=(12, 6))

# Low standard deviation plot
plt.subplot(1, 2, 1)
plt.hist(bp_readings1, bins=30, color='Green', alpha=0.7, label='Low Std Dev')
plt.axvline(mean_bp1, color='red', linestyle='dashed', linewidth=1, label='Mean (120)')
plt.title('Group A (Diet & Exercise)')
plt.xlabel('Total Cholesterol Level (mg/dL)')
plt.ylabel('Frequency')
plt.legend()

# High standard deviation plot
plt.subplot(1, 2, 2)
plt.hist(bp_readings2, bins=30, color='orange', alpha=0.7, label='High Std Dev')
plt.axvline(mean_bp2, color='red', linestyle='dashed', linewidth=1, label='Mean (120)')
plt.title('Group B (No Specific Plan))')
plt.xlabel('Total Cholesterol Level (mg/dL)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
